# Data Analysis and Visualization with python using Canada Immigration dataset.

### Introduction:
### It is important to perform some pre-operations on the dataset to make sure it is suitable for visualisation before beginning a data visualisation in Python. These pre-operations help with organising, cleaning, and getting the data ready for efficient visualisation. The pre-operations that are frequently carried out before data visualisation are described as follows:

## Process:

### 1. Import the necessary libraries: Start by importing the required libraries for data manipulation and visualization, such as Pandas, NumPy, Matplotlib, or Seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
```

### 2. Load the dataset: Load the dataset into a Pandas DataFrame. Depending on the file format (e.g., CSV, Excel, JSON), you can use specific functions like read_csv(), read_excel(), or read_json().

```python
df = pd.read_csv('./Canada.csv')
```

### 3. Data cleaning: Perform data cleaning operations to handle missing values, outliers, or inconsistencies. This process may involve dropping rows or columns with missing values (dropna()), imputing missing values with mean or median values (fillna()), or removing outliers.

#### axis=1 implies columns

```python
dp=df.dropna(how="all",axis=1)
dp
```

#### axis=0 implies columns

```python
dp=dp.dropna(how="all",axis=0)
dp
```

### Creating a copy to avoid compiler errors

```python
df_can=dp.copy()
```

### 4. To perform data analysis and enhance the convenience of working with the dataset, several steps can be taken. These steps involve modifying column names for better understanding, dropping irrelevant columns and rows, setting a specific column as the index for easy referencing, and adjusting the view of the table for improved readability

#### Droping unnecesary columns

```python
df_can.drop(['AREA', 'REG', 'DEV', 'DevName', 'Type', 'Coverage'], axis=1, inplace=True)
df_can
```

#### Renaming columns to our convinence

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)

# let's view the first five elements and see how the dataframe was changed
df_can.head()
```

#### checking wheather all coloumns are strings or not

```python

all(isinstance(column, str) for column in df_can.columns)
```

#### converting all coloumns to strings (if only false)

#### this is to avoid any data type errors in the execution

```python
df_can.columns = list(map(str, df_can.columns))

# let's check the column labels types now
all(isinstance(column, str) for column in df_can.columns)
```

#### Setting Country as an index as we visualize the data based on countries

```python
df_can.set_index('Country',inplace=True)

# let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python

# # saving the dataframe
# df_can.to_csv('file1.csv')
```

### 6. Data aggregation or grouping: Aggregating or grouping the data can be helpful for summarizing and visualizing trends or patterns. This step involves using functions like groupby(), sum(), avg(), mean(), and all other aggreation operations to aggregate data based on specific criteria.

#### Creating new column for storing the total number of immigrants from each country

```python
df_can['Total'] = df_can.sum(axis=1)

# let's check the data is updated or not
df_can.Total
```

```python
df_can
```

## Data Visualization:

### A dataset can be explored and understood better by using data visualisation. We can create various charts and visualisations that provide insights with a few queries and adding filters to the data. These visualisations help in gaining an understanding of the dataset by revealing trends, patterns, and correlations.

### In our data, we can make a area plot to track the number of immigrants over time, or a bar chart to compare immigration numbers among countries or regions. We can zoom in on the relevant subset of data for analysis by applying filters to the data, such as focusing on a specified time period.

### Lets create some charts with our refined data by adding some filters

## 1) Area plot for top 6 immigrant countries from 2000 to 2013

#### Select and store the required range of years in a new variable ('from' to 'to')

```python
years_1 = list(map(str, range(2000, 2014)))
years_1
```

#### Sort in the descending order as we required top 6 countries with highest immigrants

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top6 = df_can.head(6)
```

#### rotating the view of the table for a better understanding

```python
df_top6 = df_top6[years_1].transpose() 

df_top6
```

#### --> df.index refers to the current index of the DataFrame df. The index is a unique identifier for each row in the DataFrame.
#### --> df.index.map(int) applies the int function to each element in the index. This converts the index values to integers.
#### --> df.index = df.index.map(int) assigns the new integer values back to the index of the DataFrame.

```python
# let's change the index values of df_top6 to type integer for plotting
df_top6.index = df_top6.index.map(int)
```

#### In an area chart, the "stacked" parameter controls whether the different series or categories are stacked on top of each other or not.

```python
df_top6.plot(kind='area', stacked=False, figsize=(20, 10))

plt.title('Immigration Trend of Top 6 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

### Insights we gained from the above visualization:

#### 1. The area plot reveals that India and China had the highest number of immigrants in 2005, which then decreased but remained consistent over the years. 
#### 2. The Philippines showed a steadily increasing trend until 2005 and experienced a significant surge in immigration in 2010. 
#### 3. Other countries displayed relatively stable immigration levels without significant variations compared to India, China, and the Philippines.

#### 4. Overall, the area plot provides a clear visual representation of the immigration trends for Top 6 countires with highest immigrants to canada. It highlights the unique patterns exhibited by each country, such as the consistent levels of immigration in India and China, the increasing trend in the Philippines, and the relatively stable patterns in other countries.

## 2) A year-wise immigrant bar chart from India to Canada during the period of 1980 to 2013

#### --> range(1980, 2014) generates a sequence of integers starting from 1980 and ending at 2013 (2014 is excluded).
#### --> map(str, range(1980, 2014)) applies the str function to each element in the range. This converts each integer into a string representation.
#### --> list(map(str, range(1980, 2014))) converts the mapped elements into a list of strings.

```python
years_2=list(map(str, range(1980, 2014)))
```

```python
df_india = df_can.loc['India', years_2]
df_india
```

```python
df_india.plot(kind='bar', figsize=(13, 6))

plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('India immigrants to Canada from 1980 to 2013')

plt.show()
```

### Insights from the above bar graph:

#### In the bar graph, the highest bar is observed in the year 2005, indicating a peak in the number of Indian immigrants during that period. Similarly, another significant peak is seen in the year 2010, representing another period of high immigration from India to Canada.

#### On the other hand, the year 1985 shows the least number of Indian immigrants. Comparatively, the year 1980 has the highest number of immigrants among the years 1980 to 1986, suggesting a relatively higher number of immigrants during that initial year.

#### Overall, the bar graph provides a visual representation of the immigration trends from India to Canada over the years, highlighting the notable peaks in 2005 and 2010, as well as the contrast between the years 1980 and 1985 in terms of immigrant numbers.

## 3) A boxplot for Indian, Phillipins and China immigrants.

#### The .transpose() method is used to transpose the rows and columns of a DataFrame

```python
df_CI = df_can.loc[['China','India','Philippines'],years_2].transpose()
df_CI
```

```python
df_CI.plot(kind='box',
          figsize=(6,4))
plt.title('Box plots of Immigrants from China and India (1980 - 2013)')
plt.xlabel('Number of Immigrants')
plt.show()
```

### As reading the box plot is somewhat tricky here is the detailed explanation of the above box plot

#### 1) It displays the statistical distribution of the number of immigrants from China, India,and the Philippines over a period of 33 years, from 1980 to 2013.

#### 2) Each box plot provides a visual representation of the five-number summary of the data set for each country: the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum.

#### 3) The "whiskers" or lines extending vertically from the boxes indicate variability outside the upper and lower quartiles, hence they include the minimum and maximum data values, excluding any outliers which are represented by dots.

#### 4) The graph provides a visual comparison of immigration trends from these three countries. It allows for the observation of similarities and differences in the distribution and spread of the number of immigrants from each country over the specified period.

## 4) Finding total number of immigrants from India, China, and Philippines using Pie chart. 

#### .loc is used to access the required column, rows or specifc element in the data frame

```python
df_IF = df_can.loc[['India', 'China', 'Philippines'],'Total']
df_IF
```

#### The autopct parameter in a pie chart is used to format and display the percentage value associated with each wedge of the pie. It stands for "auto percentage" and allows you to control the formatting of the percentage value that appears on the pie chart.
#### eg: '%1.1f%%': Displays the percentage with one decimal place.
#### '%d%%': Displays the percentage as an integer value.
#### '%0.2f%%': Displays the percentage with two decimal places.

```python
df_IF.plot(kind='pie', figsize=(5, 5), autopct='%1.1f%%')

plt.title('Immigration to Canada from India and China [1980 - 2013]')

plt.show()
```

### Insights from piechart:

#### 1) The pie chart represents the distribution of immigrants from 1980 to 2013, divided into three parts: India, China, and the Philippines. 
#### 2) India has the highest percentage of immigrants, followed by China and the Philippines. When considering the total number of immigrants over the entire period, India has the highest number, followed by China. 
#### 3) The pie chart provides a clear and concise visual representation, making it easy to understand the data and its proportions.
